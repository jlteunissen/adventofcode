```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from collections import defaultdict
```

```python
dataraw = """\
jqt: rhn xhk nvd
rsh: frs pzl lsr
xhk: hfx
cmg: qnr nvd lhk bvb
rhn: xhk bvb hfx
bvb: xhk hfx
pzl: lsr hfx nvd
qnr: nvd
ntq: jqt hfx bvb xhk
nvd: lhk
lsr: lhk
rzs: qnr cmg lsr rsh
frs: qnr lhk lsr
"""
```

```python
with open("input.txt") as f:
    dataraw = f.read()
```

```python
data = defaultdict(list)
for dat in dataraw[:-1].split('\n'):
    node1, nodes = dat.split(':')
    nodes = nodes.split()
    data[node1] = nodes
```

```python
G = nx.Graph()
for key, values in data.items():
    for value in values:
        G.add_edge(key, value)
```

```python
nx.draw_spring(G, with_labels=True)
```

```python
G.remove_edge('gqr', 'vbk')
G.remove_edge('sdv', 'mxv')
G.remove_edge('klj', 'scr')
```

```python
G
```

```python
# Find connected components in the graph
connected_components = list(nx.connected_components(G))

# Print the groups and their sizes
for idx, group in enumerate(connected_components, 1):
    print(f"Group {idx}: Nodes {group}, Size {len(group)}")
```
